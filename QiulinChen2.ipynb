### Cleaning Job Data

```python
import pandas as pd
jobdf = pd.read_csv('jobdf.csv')
jobdf
```

```python
# Some faculty members did not have linkedIn and selenium might get some random wrong people.
# In order to examine whether the most uptodate position of those people are professor, I made an indicator.

jobdf = jobdf.drop('Unnamed: 0',1)
ind = [1]
for i in range(0,167,1):
    if jobdf['Name'][i] == jobdf['Name'][i+1]:
        ind.append(0)
    else:
        ind.append(1)
jobdf['i'] = ind
jobdf
```

```python
# Some faculty members did not have linkedIn and selenium might get some random wrong people.

truefac = jobdf[jobdf['i'] == 1][jobdf['Profile_url'].str.contains("rofessor")][['Name','i']]
```

```python
# Keep the true faculty group

data = jobdf.merge(truefac, on = 'Name', how = 'left')
job = data[data['i_y'].notna()]
job
```

```python
# Get a sense of the messiness...

job['Profile_url'].str.split('\n', expand=True).tail(20)
```

```python
# Split from company and employed date

company = []
emp_date = []
for i in job.index:
    a = job['Profile_url'][i].find('Company Name')
    b = job['Profile_url'][i].find('Dates Employed')
    c = job['Profile_url'][i].find('Employment Duration')
    company.append(job['Profile_url'][i][a:b])
    emp_date.append(job['Profile_url'][i][b:c])

job['Company'] = company
job['Date'] = emp_date
```

```python
# Keep non-academia activities

job = job[job['Company']!='']
job = job[~job['Company'].str.contains('niversity|MIT|UW|UC|nstitute|niversiteit|ashington')]
```

```python
company = job['Company'].str.split('\n', expand = True)[[1]]
date = job['Date'].str.split('\n', expand = True)[[1]]
job = job[['Name']].join(company)
job.columns = ['Name','Company']
job
```

```python
job = job.join(date)
job.columns = ['Name','Company','Date']
job
```

```python
# Deal with the start and end dates

start_date = []
end_date = []
for i in job.index:
    a = job['Date'][i].find('–')
    b = a - 5
    start_date.append(job['Date'][i][b:a-1])
    end_date.append(job['Date'][i][-4:])
    
job['Start'] = start_date
job['End'] = end_date
job
```

```python
# Fix type

job['Start'] = job['Start'].astype(int)
job = job[job['Start']>=2000]
job.loc[job['End'] == 'sent', 'End'] = '2021'
job['End'] = job['End'].astype(int)
```
