### Pulished Paper

I acquired data about publications of each faculty member from [Google Scholar](https://scholar.google.com/). Specifically, I focused on number of published papers and annual number of citations from year 2000 to the present.

```python
import pandas as pd
df = pd.read_csv('uwmcsmainfac.csv')
df
```

```python
# Data pre for selenium

# Set aside the names which contain "'"
df1 = df[~df['Name'].str.contains("'")] 

# Set aside the names with middlenames
df1.drop(df1[df1['Name'] == 'Josiah Hanna'].index, inplace = True)
df1.drop(df1[df1['Name'] == 'Sharon Li'].index, inplace = True)
df1.drop(df1[df1['Name'] == 'Dieter van Melkebeek'].index, inplace = True)
df1.drop(df1[df1['Name'] == 'Amos Ron'].index, inplace = True)
df1.drop(df1[df1['Name'] == 'Gurindar Sohi'].index, inplace = True)
df1
```

```python
# Pre

!pip install selenium
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
# Pre

import os
import sys
os.path.dirname(sys.executable)
```

```python
# Fetch data from Google Scholar

browser = webdriver.Chrome()
Citation_annual = []
Name = []

for p in df1['Name']:
    browser.get('https://scholar.google.com/')
    search_query = browser.find_element_by_name('q')
    search_query.send_keys(p)
    sleep(2)
    search_query.send_keys(Keys.RETURN)
    sleep(2)
    print(p)
    browser.find_element_by_link_text(p).click()
    data = browser.find_elements_by_class_name('gsc_rsb')
    sleep(2)

    for i in data[0].find_elements_by_tag_name('a'):
        Citation_annual.append(i.get_attribute('text'))
        Name.append(p)
    sleep(5)
```

```python
citationdf = pd.DataFrame(
    {'Name': Name,
     'Citation': Citation_annual
    })
```
